```python
import pandas as pd
data = pd.read_csv("tnm2.csv",index_col=0, encoding='big5') #将index_col设置为0，即指定第一列作为DataFrame的行索引
data.head()
```

```python
print("資料(資料筆數,欄位數量): ",data.shape)
print("資料中所有欄位名稱: ",data.columns)
print("資料中是否有空值: ",data.empty)
```

```python
#篩選資料:單一條件
c1 = (data["種類"] == "公共建築")
d1 = data[c1]

print("共有",len(d1),"筆資料")
d1.head()
```

```python
a = list(set(d1['所在地理區域-鄉鎮市']))
a[0]
```

```python
setdata = []
for i in range(0,len(a)):
    temp = (data["種類"] == "公共建築") & (data["所在地理區域-鄉鎮市"] == a[i])
    dataset = data[temp]
    print(a[i],":",len(dataset))
    setdata.append(len(dataset))
```

```python
newsetdata = pd.DataFrame()
newsetdata["所在地理區域-鄉鎮市"] = a
newsetdata["區域古蹟數"] = setdata
newsetdata
```

```python
#篩選資料:多條件
c2 = (data["種類"] == "公共建築") & (data["所在地理區域-鄉鎮市"] ==  "東區")
d2 = data[c2]

print("共有",len(d2),"筆資料")
d2.head()
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.ptt.cc/bbs/Vtuber/index616.html'
web = requests.get(url) #取得網頁內容
soup = BeautifulSoup(web.text, "html.parser") #使用html.parser解析器
title = soup.title
print(title.text) #印出title的文字 
```

```python
# 找到所有文章
articles = soup.find_all('div', class_='r-ent')#'r-ent' 包含文章的各項資訊，例如標題、作者、發文時間、推文數等

# 遍歷每個文章
for article in articles:
    # 找到日期
    date = article.find('div', class_='date').text.strip()

    # 找到標題和網址  #title_element.text.strip() 是用來取得標題文字,.strip() 則是用來去除字串前後的空白字元
    title_element = article.find('div', class_='title')
    if title_element is not None:
        title = title_element.text.strip()
        url = 'https://www.ptt.cc' + title_element.find('a')['href']
    else:
        title = '無標題'
        url = ''

    # 列印出日期、標題和網址
    print(date, url, title)
```

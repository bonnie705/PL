```python
import requests
import json

# 設定搜尋位置和半徑
location = '25.026377,121.527326'  # 台師大經緯度
radius = 2000  # 兩公里

# 設定 Places API 的請求 URL 和 API 金鑰
url = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json'
api_key = 'AIzaSyDpzE4HxJSMJk6G9IK0m1aDeTjRrkPx1js'  # 記得替換成自己的 API 金鑰

# 設定搜尋類型為餐廳和飲料店
types = 'restaurant|cafe'

# 發送 Places API 的請求
response = requests.get(f'{url}?location={location}&radius={radius}&types={types}&key={api_key}')

# 解析 Places API 回傳的 JSON 格式資料
places_data = json.loads(response.text)

# 顯示搜尋結果
print(json.dumps(places_data, indent=4))
```

```python


# 取得每個地點的詳細資訊，並加入評價和分類的中文描述
for place in places_data['results']:
    # 發送 Place Details API 的請求
    details_url = 'https://maps.googleapis.com/maps/api/place/details/json'
    details_response = requests.get(f'{details_url}?place_id={place["place_id"]}&key={api_key}')
    details_data = json.loads(details_response.text)

```

```python
# 取得分類和評價的中文描述
types = details_data['result']['types']
types_desc = []
for t in types:
    type_url = 'https://maps.googleapis.com/maps/api/place/autocomplete/json'
    type_response = requests.get(f'{type_url}?types=establishment&input={t}&key={api_key}')
    type_data = json.loads(type_response.text)
    if len(type_data['predictions']) > 0:
        types_desc.append(type_data['predictions'][0]['description'])
```

```python
import requests
import json
from googletrans import Translator
from datetime import datetime

api_key = 'YAIzaSyDpzE4HxJSMJk6G9IK0m1aDeTjRrkPx1js'  # 替換為自己的 API key
place_id = 'ChIJ8eWRPYKpQjQRuXQ3tGY9wKo'  # 指定餐廳的 place_id

# 取得餐廳的詳細資訊
details_url = 'https://maps.googleapis.com/maps/api/place/details/json'
details_response = requests.get(f'{details_url}?place_id={place_id}&fields=name,formatted_address,rating,review,photo&key={api_key}')
details_data = json.loads(details_response.text)

# 取得餐廳的名稱、評分等資訊
name = details_data['result']['name']
address = details_data['result']['formatted_address']
rating = details_data['result']['rating']

# 取得評論並翻譯成中文
translator = Translator()
reviews = []
for review in details_data['result']['reviews']:
    text = review['text']
    text_translated = translator.translate(text, dest='zh-TW').text
    rating = review['rating']
    time = datetime.fromtimestamp(review['time'])
    reviews.append({
        'text': text_translated,
        'rating': rating,
        'time': time
    })

# 輸出結果
print(f'餐廳名稱：{name}')
print(f'地址：{address}')
print(f'評分：{rating}')
for review in reviews:
    print('---')
    print(f"評論：{review['text']}")
    print(f"評分：{review['rating']}")
    print(f"時間：{review['time']}")
```
